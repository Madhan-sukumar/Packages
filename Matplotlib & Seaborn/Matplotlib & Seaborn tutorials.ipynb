## Setting axis labels to the plots

```python
fig = plt.figure()
axis = fig.add_subplot(1,1,1)
```

```python
fig = plt.figure()
axis = fig.add_subplot(1,1,1)

#creating the random data
data = np.random.randn(1000).cumsum()

axis.plot(data)

#creating the x and y lables 
axis.set_xlabel('Index')
axis.set_ylabel('Value')

axis.set_title('Cummulative sum of random normal data')
plt.show()
```

## Setting ticks and ticklabels

```python
fig = plt.figure()
axis = fig.add_subplot(1,1,1)

#creating the random data
data = np.random.randn(1000).cumsum()

axis.plot(data,color = 'r')

#creating the x and y lables 
axis.set_xlabel('Index')
axis.set_ylabel('Value')

axis.set_title('Cummulative sum of random normal data')

#creating ticks for x axis
axis.set_xticks([0,250,500,750,1000])
axis.set_xticklabels(['Start','Quater','Half','Three_quater','Stop'] , 
                    rotation = 45,
                    fontsize='large')

plt.show()
```

## Adding legends to a plot

```python
x1 = np.random.randn(1000).cumsum()
x2= np.random.randn(1000).cumsum()
x3= np.random.randn(1000).cumsum()

fig,axis = plt.subplots(1,1)

axis.plot(x1,color='r',label='x1')
axis.plot(x2,color='g',label='x2')
axis.plot(x3,color='y',label='x3')

#creating the x and y lables 
axis.set_xlabel('X')
axis.set_ylabel('Value')

axis.set_title('cumsum plots')

#creating ticks for x axis
axis.set_xticks([0,250,500,750,1000])
axis.set_xticklabels(['Start','Quater','Half','Three_quater','Stop'] , 
                    rotation = 45,
                    fontsize='large')

#creating legends
axis.legend(loc='best')

plt.show()

```

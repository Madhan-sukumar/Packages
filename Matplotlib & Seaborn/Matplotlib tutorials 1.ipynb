### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Line plot

```python
#creating a 1d array
x = [4,5,6,7,8]
y = [6,8,3,4,7]
```

```python
plt.plot(x,y)

plt.title('x-y line graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show
```

```python
#creating a points in line
plt.plot(x,y)
plt.plot(x,y,'^')

plt.title('x-y line graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show
```

## comparison plots

```python
#creating two 1d array
x = [2,4,6,8,9]
y = [1,2,5,7,9]

x1=[1,2,3,6,8]
y1=[2,3,5,8,9]
```

```python
plt.plot(x,y, color='c',linewidth=4)
plt.plot(x1,y1,color='m',linewidth=2)

plt.title('Comparison Plots')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#annotatimg the graph
plt.annotate(xy=[2.2,3],text='Second line')
plt.annotate(xy=[4.5,2],text='First line')

#creating a reference box
plt.legend(['First line','Second line'],loc=4)
plt.show
```

## BAR GRAPH

### Creating vertical bar graph using bar function

```python
#creating two 1d array
x = [1,2,4,5,6]
y = [6,7,8,9,12]

x1=[2,5,7,9,10]
y1=[5,7,9,10,12]

plt.bar(x,y, color='c',linewidth=4)
plt.bar(x1,y1,color='m',linewidth=2)

plt.title('Comparison Plots')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#annotatimg the graph
plt.annotate(xy=[2.2,3],text='Second line')
plt.annotate(xy=[4.5,2],text='First line')

#creating a reference box
plt.legend(['First line','Second line'],loc=2)
plt.show
```

### Creating horizontal bar graph using barh function

```python
#creating two 1d array
x = [1,2,4,5,6]
y = [6,7,8,9,12]

x1=[2,5,7,9,10]
y1=[5,7,9,10,12]

plt.barh(x,y, color='c',linewidth=4)
plt.barh(x1,y1,color='m',linewidth=2)

plt.title('Comparison Plots')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#annotatimg the graph
plt.annotate(xy=[2.2,3],text='Second line')
plt.annotate(xy=[4.5,2],text='First line')

#creating a reference box
plt.legend(['First line','Second line'],loc=4)
plt.show
```

# BUBBLE & PIE CHART

### Bubble graph

```python
#creating a 1d array
x = [4,5,6,7,8]
y = [6,8,3,4,7]
```

```python
# creating using scatter function and s&c variable used to mention size&colour and generating randomly.
plt.scatter(x,y, s = np.random.rand(5)*500,c=np.random.rand(5))
plt.title('Scatter graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show
```

### Pie chart

```python
data = [2,6,8,3,4]

#creating explode for a particular pie
ex=(0,0,0.1,0,0)
lb=('A','B','C','D','E')
plt.pie(data,explode=ex,labels=lb,shadow=True)
plt.show
```

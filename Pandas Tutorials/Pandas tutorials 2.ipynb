## Merging two dataframes using joins

```python
#creating a df using dict for leftdataframe
l_df = {'Student':['Madhan','Surya','Tom','Jake','Murali'],
                     'Age':[23,24,25,26,27],
                     'Score_1':[35,86,36,96,83],
                     'Score_2':[74,84,24,57,94],
                     'Score_3':[64,43,74,63,85],
                     'Score_4':[56,36,85,84,34]
    
}

left_df = pd.DataFrame(l_df)
left_df
```

```python
#creating a df using dict for right dataframe
r_df = {'Student':['Madhan','Surya','Tom'],
                     'Age':[23,24,25],
                     'Score_5':[75,66,35],
                     'Score_6':[74,84,64],
                     
}

right_df = pd.DataFrame(r_df)
right_df
```

### Inner join

```python
merged_df = pd.merge(left_df,right_df)
```

```python
#gives only the common value rows(Age) between two df
merged_df
```

### Outer join

```python
#creating a df using dict for leftdataframe
l_df = {'Student':['Madhan','Surya','Tom','Jake','Murali'],
                     'Age':[23,24,25,26,27],
                     'Score_1':[35,86,36,96,83],
                     'Score_2':[74,84,24,57,94],
                     'Score_3':[64,43,74,63,85],
                     'Score_4':[56,36,85,84,34]
    
}

left_df = pd.DataFrame(l_df)
left_df
```

```python
#creating a df using dict for right dataframe
r_df = {'Student':['Madhan','Surya','Tom'],
                     'Age':[28,26,26],
                     'Score_5':[75,66,35],
                     'Score_6':[74,84,64],
                     
}

right_df = pd.DataFrame(r_df)
right_df
```

```python
#merged both the dataframe unless it shouldn't have common values
merged_df = pd.merge(left_df,right_df,how="outer")
merged_df
```

```python
#merged both the dataframe based on keys
merged_df = pd.merge(left_df,right_df,how="outer",on=["Student", "Age"])
merged_df
```

```python
#merged both the dataframe based on keys
#creates a different column for age
merged_df = pd.merge(left_df,right_df,how="outer",on=["Student"])
merged_df
```

### Left join

```python
merged_df = pd.merge(left_df,right_df,how="left")
merged_df
```

### Right join

```python
merged_df = pd.merge(left_df,right_df,how="right")
merged_df
```

## Merging using `.join()` function

```python
#joining based on left in outer join
left_df.join(right_df,lsuffix="_left")
```

```python
#joining based on right in outer join
right_df.join(left_df,lsuffix = "_right",rsuffix="_left")
```

## Cumulative sum and cumulative product

```python
#creating a df using dict for leftdataframe
cs_df = {'Student':['Madhan','Surya','Tom','Jake','Murali'],
                     'Age':[23,24,25,26,27],
                     'Score_1':[35,86,36,96,83],
                     'Score_2':[74,84,24,57,94],
                     'Score_3':[64,43,74,63,85],
                     'Score_4':[56,36,85,84,34]
    
}

cus_df = pd.DataFrame(cs_df)
cus_df
```

```python
#gave the cummulative sum of score1 column
cus_df['Score_1'].cumsum()
```

```python
#cumulative product of score2 column
cus_df['Score_2'].cumprodd()
```

```python
# cumulative sum on series of columns where sub1 score added to score2, score2 added to score3 only if axis=1
cus_df.iloc[:,2:].cumsum(axis=1)
```

```python
# cumulative sub based on rows on all columns
cus_df.iloc[:,2:].cumsum()
```

## Checking and Droping the null values

### Checking the null values

```python
#creating a df using dict 
nu_df = {'Student':['Madhan','Surya','Tom','Jake','Murali'],
                     'Age':[23,24,25,26,27],
                     'Score_1':[35,86,36,96,83],
                     'Score_2':[74,84,24,57,94],
                     'Score_3':[64,43,74,63,85],
                     'Score_4':[56,36,85,84,34]
    
}

null_df = pd.DataFrame(nu_df)
null_df
```

```python
# now creating a column having null values
null_df['Score_5'] = [54,74,np.nan,65,None]
```

```python
null_df
```

```python
null_df.isnull()
```

```python
#another method
null_df.isna()
```

```python
null_df.info()
```

```python
null_df.isnull().sum()
```

```python
# to see the null value rows
null_df[null_df['Score_5'].isnull()].sample(2)
```

## Dropping the null values from Dataframe

```python
#droped the null vaulues as rows
droped_df = null_df.dropna(axis=0)
droped_df
```

```python
#dropped the null values as columns
droped_df1 = null_df.dropna(axis=1)
droped_df1
```

## Filling the null values

```python
#filling the null values as zero
filled_df1 = null_df.fillna(0)
filled_df1
```

```python
#filling the null values with mean values
filled_df2 = null_df.fillna(null_df['Score_5'].mean())
filled_df2
```

```python
#filling the null values with median values
filled_df3 = null_df.fillna(null_df['Score_5'].median())
filled_df3
```

## Grouping the data

```python
#creating a df using dict
gp_df = pd.DataFrame({'Student':['Madhan','Surya','Tom','Madhan','Tom','Murali'],
                     'Age':np.random.randint(20,25,size=6),
                     'Score_1':np.random.randint(50,100,size=6),
                     'Score_2':np.random.randint(70,100,size=6),
                     'Score_3':np.random.randint(40,90,size=6),
                     'Score_4':np.random.randint(40,80,size=6),
                     'Score_5':np.random.randint(50,100,size=6),
                     'Score_6':np.random.randint(30,70,size=6),
    
})
gp_df
```

```python
# Grouping the data bsed on key
#now it groups the data but now in the visibility
gp_df.groupby('Student')
```

```python
gp_df.groupby('Student').sum()
```

```python
gp_df.groupby('Student').mean()
```

```python
#grouping by cumulative sum
gp_df.groupby('Student').cumsum()
```

```python
#grouping by standard deviation
gp_df.groupby('Student').std()
```

```python
gp_df.groupby('Student').count()
```

## Grouping based on more that one column


```python
#creating a df using dict
gp1_df = pd.DataFrame({'Student':['Madhan','Surya','Tom','Madhan','Tom','Murali'],
                     'Age':np.random.randint(20,25,size=6),
                    'Semester' : [1,1,1,2,2,1],
                     'Score_1':np.random.randint(50,100,size=6),
                     'Score_2':np.random.randint(70,100,size=6),
                     'Score_3':np.random.randint(40,90,size=6),
                     'Score_4':np.random.randint(40,80,size=6),
                     'Score_5':np.random.randint(50,100,size=6),
                     'Score_6':np.random.randint(30,70,size=6),
    
})
gp1_df
```

```python
#grouping based on key from two columns( Student & semester)
gp1_df.groupby([gp1_df['Student'],gp1_df['Semester']]).sum()
```

```python
# grouping the data based on key for 1 column such as score_3 column alone
gp1_df.groupby('Student')['Score_3'].sum()
```

## Sorting the data

```python
#creating a df using dict
so_df = pd.DataFrame({'Student':['Madhan','Surya','Tom','Madhan','Murali'],
                     'Age':np.random.randint(20,25,size=5),
                     'Score_1':np.random.randint(50,100,size=5),
                     'Score_2':np.random.randint(70,100,size=5),
                     'Score_3':np.random.randint(40,90,size=5),
                     'Score_4':np.random.randint(40,80,size=5),
                     'Score_5':[74,23,np.nan,75,None]
                     
    
})
so_df
```

### Shuffling the dataframe

```python
# atfirst we shuffling the dataframe before sorting
#see the index , where the data got shuffled
so_df = so_df.sample(frac=1)
so_df
```

### Sorting the dataframe

#### Sorting the dataframe in rows

```python
#sorting the dataframe by index
so_df = so_df.sort_index()
so_df
```

```python
# sorting the dataframe based on age column in ascending
so_df = so_df.sort_values(by='Age')
so_df
```

```python
# sorting the dataframe based on age column in descending
so_df = so_df.sort_values(by='Age',ascending = False)
so_df
```

#### Sorting the dataframe in columns

```python
#sorting the dataframe in columns alphabetically as ascending  
so_df.sort_index(axis=1)
```

```python
#sorting the dataframe in columns alphabetically as descening
so_df.sort_index(axis=1,ascending=False)
```

```python
# sorting the dataframe based on rows in specific column
#ordered alphabetically in student column
so_df.sort_values(by='Student')
```
